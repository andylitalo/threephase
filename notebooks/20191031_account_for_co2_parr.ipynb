# Accounting for CO2 in Parr Reactor

Initially, I believed that I could estimate the amount of CO2 in the Parr reactor simply by using the CO2 equation of state to estimate the amount dispensed by the ISCO pump into the Parr reactor. This method clearly overestimates the actual amount of CO2 in the Parr reactor because a substantial amount of CO2 leaked during the experiment and I am not confident that the ISCO pump was always liquid full when refilled with the liquid CO2 tank. Therefore, I tried two other methods to estimate the amount of CO2 in the Parr reactor. The first uses the change in density of CO2 in the vapor phase of a CO2-C5 binary coexistence at the pressure and temperature before and after adding CO2, assuming that CO2 and C5 equilibrate immediately after adding CO2 or C5 to the Parr reactor or venting and that no CO2 enters the liquid polyol-rich phase. The second uses the PC-SAFT model developed by Dr. Huikuan Chao to estimate the composition of the vapor phase, but this circularly uses the model for the validation data, and therefore is only intended to be a rough estimate. In both of the latter cases, the estimates suffer from not accounting for the possible presence of a third phase, which I believe phase separates in the middle of the experiment and might have a substantial effect on the composition of the liquid and vapor phases.

## ISCO Estimate

The first method I used to estimate the mass of CO2 in the Parr reactor is estimating how much CO2 was dispensed from the ISCO pump into the Parr reactor based on the equation of state of CO2. I recorded the volume and pressure recorded by the ISCO pump both before and after injection of CO2 into the Parr reactor, and I assumed that the temperature remained constant at the lab temperature (about 21 C). I then used the equation of state of CO2 to estimate the beginning and final masses of CO2 in the ISCO, using the difference as an estimate for the amount injected into the Parr reactor. 

### Assumptions

This method assumes that 
1. The ISCO pump is liquid-full of CO2 and therefore contains a single, homogeneous phase of CO2 at all times
2. The Parr reactor does not leak
3. The temperature of the ISCO pump is constant and homogeneous
4. The pressure transducer of the ISCO pump does not drift

The first assumption is certainly false when the ISCO pump is refilled because the liquid CO2 from the tank must expand to fill the dead volume, and the pressure of the tank is not sufficient to re-condense that vaporized CO2. Whether the CO2 became homogeneously liquid when pressurized to 1000 psi and above before injection is not clear to me right now and should be tested using our ISCO pump.

The second assumption is definitely false because I felt vapor leak through a needle valve on the gas-sampling port of the Parr reactor. I have estimated the amount of CO2 that leaks based on differences in pressure betIen injections of CO2 and C5, but I do not have any way to validate these estimates.

The third (3) and fourth (4) assumptions are fairly robust, as the steel syringe of the ISCO pump conducts heat Ill enough to maintain thermal equilibrium with the laboratory and periodic checks of the pressure transducer reading when emptying the ISCO or loading with the liquid CO2 tank at a known pressure did not show signs of drift beyond 10 psi, which would have a negligible effect on the estimated amount of CO2 injected in the Parr reactor.

Overall, I believe that this estimate is an *overestimate* of the true mass of CO2 in the Parr reactor due to the flimsiness of assumptions (1) and (2).

## CO2-C5 Binary Coexistence Estimate

This method assumes that because the amount of polyol in the vapor phase is negligible, I can treat the vapor phase as the vapor phase of a CO2-C5 binary coexistence. Under this assumption, I use our PC-SAFT model (with parameters fitted to Eckert and Sandler's data (1986)) to compute the the vapor-liquid equilibrium of CO2 and C5 at the pressure and temperature in the Parr reactor both before injecting with the ISCO and immediately after. I then take the difference in density of CO2 in the vapor phase and multiply by the estimated volume of the vapor phase, which I estimate by subtracting the estimated liquid volume from the approximate interior volume of the Parr reactor (1200 mL). I estimate the liquid volume as $V_{liq} = m_{poly} / \rho_{poly}^{HPLIS}$, where $\rho_{poly}^{HPLIS} = \rho_{poly}^{atm}(T) v_{poly}^{HPLIS}$, where $\rho_{poly}^{atm}(T)$ is the estimated density of polyol under atmospheric pressure at the given temperature and $v_{poly}^{HPLIS}$ is the volumetric fraction of polyol in the HPLIS, inferred by estimating the volumes of CO2 and C5 based on their masses measured by the gas chromatograph and their approximate densities at the given temperature.

### Assumptions

This method makes the following assumptions:
1. There is no polyol in the vapor phase
2. The vapor-liquid equilibrium between CO2 and C5 is achieved very rapidly (minutes)
3. The vapor-liquid equilibrium between CO2 and C5 is not affected by polyol in the liquid phase (e.g. the polyol does not enhance adsorption of CO2 into the liquid phase)
4. No third phase forms
5. Eckert and Sandler's data (1986) yield accurate PC-SAFT parameters for the binary coexistence

The first assumption is likely valid because of the low vapor pressure of polyol given its molecular weight of 1000 g/mol and surface tension of almost 30 mN/m. A quick sniff assures the scientist that this is indeed the case.

The second third, and fourth assumptions are likely inaccurate. The vapor-liquid equilibrium will definitely be affected by the presence of polyol in the liquid as this will lower the diffusivity (hindering equilibrium between vapor-phase and liquid-phase CO2 and C5) and will affect solubility in the liquid phase. This is made clear when the PC-SAFT estimates of C5 weight fraction in the vapor phase do not match the GC estimates. The fourth assumption is likely false by the sixth injection based on preliminary evidence of the formation of a third phase then, and it is likely that the third phase is present in later measurements as well.

The fourth assumption is probably trustworthy since the data are plentiful and precise.

The advantage of this method is that it only considers changes in CO2 mass on the order of a few minutes, so we can neglect the leaking of CO2 and actually use this method as an estimate for how much Co2 leaked between injections from the ISCO. 

## CO2-C5-Polyol Ternary Coexistence Estimate

This final method uses the PC-SAFT model for the ternary coexistence among polyol, CO2, and C5 that we intend to validate with these data to estimate the mass of CO2. This is obviously circular reasoning, but as a sanity check I wanted to see what values it predicted. In general, the estimates were within 30% of the CO2-C5 binary coexistence values but occasionally varied more substantially. They did not agree at all with the ISCO method estimates.

### Assumptions

This method makes the following assumptions:
1. The PC-SAFT model for the ternary system is accurate, although C5-polyol coexistence data are missing and parameters are estimated based on CO2-polyol coexistence
2. No third phase forms

Both of these assumptions are poor, the first because the model is validated by the predictions being made and the second because we have evidence that the third phase forms. This second assumption might be why the estimates of CO2 mass are much lower than those estimated by the other methods.

##  Leak Estimate

We estimate the leak of CO2 using the CO2-C5 binary coexistence method's estimate of the mass of CO2. Assuming that changes in the estimated mass are only due to the leaking of CO2 or injections of CO2 from the ISCO pump, we take the difference in mass (in general a decrease) between injections from the ISCO and divide by the elapsed time to estimate the rate of leak of CO2. We notice an increase in the rate later in the experiment, around the time that we began to notice leaking through the needle valve on the gas-sampling port of the Parr reactor.

## Plot of CO2 Mass Over Time

Below, I plot the estimated mass of CO2 over time for both the ISCO and CO2-C5 binary coexistence methods. I do not plot the estimate from the ternary PC-SAFT model due to the circularity of the method, although I note that the values are slightly loIr than those from the CO2-C5 estimate. I also plot the estimated amount of CO2 leaked, which is based on changes in the estimated mass of CO2 betIen injections from the ISCO pump using the CO2-C5 binary coexistence (so the leak rate is likely erroneous). Finally, I plot the "corrected" masses, which have subtracted the estimated mass of leaked CO2.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import plot
import tern


# data filepath
data_filepath = '..//ternary_system_results//20191002_gc_data_proc_corr.csv'

# load data
df = pd.read_csv(data_filepath)
t = df['elapsed time [min]'].to_numpy(dtype=float)
m_co2_isco = df['mass co2 [g]'].to_numpy(dtype=float)
m_co2_bin = df['mass co2 [g] (bin)'].to_numpy(dtype=float)
m_co2_leak = df['mass co2 (leak) [g]'].to_numpy(dtype=float)
m_co2_isco_corr = df['mass co2 [g] (corr)'].to_numpy(dtype=float)
m_co2_bin_corr = df['mass co2 [g] (bin) (corr)'].to_numpy(dtype=float)
```

Next, I will plot the data to compare the different methods for estimating the mass of CO2

```python
ax = plot.plot_line(t, m_co2_isco, xlabel='time [min]', ylabel=r'$m_{CO2}$ [g]', 
                    title='Comparison of Methods to Estimate CO2 Mass', marker='o', lw=1, color='b', label='ISCO')
ax = plot.plot_line(t, m_co2_bin, ax=ax, marker='o', lw=1, color='g', label='CO2-C5 Coex')
ax = plot.plot_line(t, m_co2_isco_corr, ax=ax, marker='o', lw=1, color='b', label='ISCO (corr)', ls='--')
ax = plot.plot_line(t, m_co2_bin_corr, ax=ax, marker='o', lw=1, color='g', label='CO2-C5 Coex (corr)', ls='--')
ax = plot.plot_line(t, m_co2_leak, ax=ax, marker='', lw=1, color='r', label='leak', ls='--')
plt.legend(loc='best')
```

Note that only the corrected CO2-C5 coexistence estimate of the CO2 mass returns to zero at the end. This is because the leak is defined based on how much the estimated CO2 mass decreases betIen injections of CO2, such that all "excess" CO2 is considered to have leaked out. This method is likely inaccurate because I expect that our estimate of the amount of CO2 added in the first place is incorrect using the CO2-C5 coexistence (see above for reasons) and because I expect that while venting at the end, I not only vented headspace but also the liquid-like CO2-C5 third phase, meaning our estimate of the amount of CO2 vented would be too low. Nevertheless, the leak rate is very low during this venting period, which suggests that perhaps the CO2 lost from the hypothetical third phase due to venting is negligible.

Additionally, note that the estimates before about 6000 minutes are not terribly different, and before 2000 minutes are fairly close, suggesting that I can rely on our estimates of the mass of CO2 for early measurements, but not on our estimates for later measurements.
